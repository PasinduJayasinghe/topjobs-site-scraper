```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.topjobs.lk/applicant/vacancybyfunctionalarea.jsp?FA=SDQ'
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup.prettify)
```

```python
soup.find_all('tr')
```

```python
table=soup.find('table',{'class':'tbldata_2 vbfa-table'})
```

```python
rows=table.find_all('tr')
```

```python
for row in rows:
  cells=row.find_all('td')
  for cell in cells :
    roles=cell.find_all('h2')
    for role in roles:
      print(role.text)
    companies=cell.find_all('h1')
    for company in companies:
      print(company.text)

```
